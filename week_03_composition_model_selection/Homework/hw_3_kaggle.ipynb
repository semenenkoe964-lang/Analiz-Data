```python
plt.rcParams['figure.figsize'] = [15, 12]
plt.rcParams['font.size'] = 10

fig_num, axes_num = plt.subplots(2, 3, figsize=(18, 10))
fig_num.suptitle('  ', fontsize=16, fontweight='bold')

for idx, feature in enumerate(num_cols):
    # 
    ax_hist = axes_num[0, idx]
    ax_hist.hist(data[feature].dropna(), bins=30, edgecolor='black', alpha=0.7, color='skyblue')
    ax_hist.set_title(f': {feature}', fontsize=12, fontweight='bold')
    ax_hist.set_xlabel(feature)
    ax_hist.set_ylabel('')
    ax_hist.grid(True, alpha=0.3)

    # Boxplot
    ax_box = axes_num[1, idx]
    boxplot_data = ax_box.boxplot(data[feature].dropna(), patch_artist=True)
    boxplot_data['boxes'][0].set_facecolor('lightcoral')
    boxplot_data['whiskers'][0].set_color('darkred')
    boxplot_data['whiskers'][1].set_color('darkred')
    boxplot_data['caps'][0].set_color('darkred')
    boxplot_data['caps'][1].set_color('darkred')
    boxplot_data['medians'][0].set_color('darkred')
    ax_box.set_title(f'Boxplot: {feature}', fontsize=12, fontweight='bold')
    ax_box.set_ylabel(feature)
    ax_box.grid(True, alpha=0.3)

plt.tight_layout()
plt.subplots_adjust(top=0.93)


#  9      
selected_categorical = ['Sex', 'HasPartner', 'HasChild', 'HasInternetService',
                       'HasContractPhone', 'IsBillingPaperless', 'PaymentMethod',
                       'Churn', 'HasOnlineSecurityService']

fig_cat, axes_cat = plt.subplots(3, 3, figsize=(18, 15))
fig_cat.suptitle('  ', fontsize=16, fontweight='bold')

axes_cat = axes_cat.flatten()

for idx, feature in enumerate(selected_categorical):
    if feature in data.columns:
        #  
        value_counts = data[feature].value_counts()

        #  subplot     bar  pie
        ax_bar = axes_cat[idx]

        # Bar 
        bars = ax_bar.bar(value_counts.index.astype(str), value_counts.values,
                         color=plt.cm.Set3(np.arange(len(value_counts))))
        ax_bar.set_title(feature, fontsize=12, fontweight='bold')
        ax_bar.set_xlabel('')
        ax_bar.set_ylabel('')
        ax_bar.tick_params(axis='x', rotation=45)

        #    
        for bar, value in zip(bars, value_counts.values):
            height = bar.get_height()
            ax_bar.text(bar.get_x() + bar.get_width()/2., height + max(value_counts.values)*0.01,
                       f'{value}\n({value/len(data)*100:.1f}%)',
                       ha='center', va='bottom', fontsize=9)

        ax_bar.grid(True, alpha=0.3)

#   subplots    9
for idx in range(len(selected_categorical), len(axes_cat)):
    fig_cat.delaxes(axes_cat[idx])

plt.tight_layout()
plt.subplots_adjust(top=0.95)



remaining_features = [f for f in cat_cols if f not in selected_categorical]
if remaining_features:
    print(f"\n  ({len(remaining_features)}): {', '.join(remaining_features)}")

    #      
    n_remaining = len(remaining_features)
    n_cols = 3
    n_rows = (n_remaining + n_cols - 1) // n_cols

    if n_rows > 0:
        fig_extra, axes_extra = plt.subplots(n_rows, n_cols, figsize=(15, 4*n_rows))
        fig_extra.suptitle('  ', fontsize=16, fontweight='bold')

        if n_rows == 1:
            axes_extra = axes_extra.reshape(1, -1)

        for idx, feature in enumerate(remaining_features):
            row = idx // n_cols
            col = idx % n_cols

            if feature in data.columns:
                value_counts = data[feature].value_counts()

                #  
                ax = axes_extra[row, col]
                wedges, texts, autotexts = ax.pie(value_counts.values,
                                                  labels=value_counts.index.astype(str),
                                                  autopct='%1.1f%%',
                                                  startangle=90,
                                                  colors=plt.cm.Pastel1(np.arange(len(value_counts))))
                ax.set_title(feature, fontsize=12, fontweight='bold')


        #   subplots
        for idx in range(len(remaining_features), n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            fig_extra.delaxes(axes_extra[row, col])

        plt.tight_layout()
        plt.subplots_adjust(top=0.93)

plt.show()
```
