```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
import warnings

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# –ö–æ–Ω—Å—Ç–∞–Ω—Ç—ã
RANDOM_STATE = 42
TEST_SIZE = 0.2
N_ESTIMATORS = 100


def load_data():
    """–ó–∞–≥—Ä—É–∑–∫–∞ –∏ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö."""
    train_data = pd.read_csv('train.csv')
    test_data = pd.read_csv('test.csv')
    sample_submission = pd.read_csv(
        'gender_submission.csv'
    )

    return train_data, test_data, sample_submission


def explore_data(train_df, test_df):
    """–ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö."""
    print("=" * 60)
    print("–î–ê–ù–ù–´–ï –î–õ–Ø –û–ë–£–ß–ï–ù–ò–Ø:")
    print("=" * 60)
    print(f"–†–∞–∑–º–µ—Ä: {train_df.shape}")
    print(f"\n–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö:")
    print(train_df.info())

    print(f"\n–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
    print(train_df.isnull().sum())

    print(f"\n–û—Å–Ω–æ–≤–Ω—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏:")
    print(train_df.describe())

    print("\n" + "=" * 60)
    print("–¢–ï–°–¢–û–í–´–ï –î–ê–ù–ù–´–ï:")
    print("=" * 60)
    print(f"–†–∞–∑–º–µ—Ä: {test_df.shape}")
    print(f"\n–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
    print(test_df.isnull().sum())

    # –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
    survival_counts = train_df['Survived'].value_counts()
    survival_percent = train_df['Survived'].value_counts(normalize=True) * 100

    print("\n" + "=" * 60)
    print("–†–ê–°–ü–†–ï–î–ï–õ–ï–ù–ò–ï –¶–ï–õ–ï–í–û–ô –ü–ï–†–ï–ú–ï–ù–ù–û–ô:")
    print("=" * 60)
    for i, (count, percent) in enumerate(zip(survival_counts, survival_percent)):
        status = "–í—ã–∂–∏–ª" if i == 1 else "–ü–æ–≥–∏–±"
        print(f"{status}: {count} ({percent:.1f}%)")

    return survival_counts
```

```python
def feature_engineering(df):
    """–°–æ–∑–¥–∞–Ω–∏–µ –∏ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤."""
    df_processed = df.copy()

    # 1. –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ –∏–º–µ–Ω–∏
    df_processed['Title'] = df_processed['Name'].str.extract(
        ' ([A-Za-z]+)\.',
        expand=False
    )

    # –ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ —Ä–µ–¥–∫–∏—Ö —Ç–∏—Ç—É–ª–æ–≤
    title_mapping = {
        'Mr': 'Mr',
        'Miss': 'Miss',
        'Mrs': 'Mrs',
        'Master': 'Master',
        'Dr': 'Rare',
        'Rev': 'Rare',
        'Col': 'Rare',
        'Major': 'Rare',
        'Mlle': 'Miss',
        'Countess': 'Rare',
        'Ms': 'Miss',
        'Lady': 'Rare',
        'Jonkheer': 'Rare',
        'Don': 'Rare',
        'Dona': 'Rare',
        'Mme': 'Mrs',
        'Capt': 'Rare',
        'Sir': 'Rare'
    }

    df_processed['Title'] = df_processed['Title'].map(title_mapping)

    # 2. –°–µ–º–µ–π–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
    df_processed['FamilySize'] = (
        df_processed['SibSp'] + df_processed['Parch'] + 1
    )
    df_processed['IsAlone'] = (df_processed['FamilySize'] == 1).astype(int)

    # 3. –ü—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –∫–∞—é—Ç—ã
    df_processed['Deck'] = df_processed['Cabin'].str[0]
    df_processed['Deck'] = df_processed['Deck'].fillna('Unknown')

    # 4. –í–æ–∑—Ä–∞—Å—Ç–Ω—ã–µ –≥—Ä—É–ø–ø—ã
    df_processed['AgeGroup'] = pd.cut(
        df_processed['Age'].fillna(df_processed['Age'].median()),
        bins=[0, 12, 18, 35, 60, 100],
        labels=['Child', 'Teen', 'YoungAdult', 'Adult', 'Senior']
    )

    # 5. –ü—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –±–∏–ª–µ—Ç–∞
    df_processed['TicketPrefix'] = df_processed['Ticket'].apply(
        lambda x: x.split()[0] if len(x.split()) > 1 else 'None'
    )
    df_processed['TicketNumber'] = df_processed['Ticket'].apply(
        lambda x: x.split()[-1]
    )
    df_processed['TicketLength'] = df_processed['TicketNumber'].apply(len)

    return df_processed


def preprocess_data(train_df, test_df):
    """–ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö."""
    # –û–±—ä–µ–¥–∏–Ω—è–µ–º –¥–ª—è —Å–æ–≥–ª–∞—Å–æ–≤–∞–Ω–Ω–æ–π –æ–±—Ä–∞–±–æ—Ç–∫–∏
    combined = pd.concat([train_df.drop('Survived', axis=1), test_df], axis=0)
    combined_reset = combined.reset_index(drop=True)

    # 1. –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    categorical_features = ['Sex', 'Embarked', 'Title', 'Deck',
                           'AgeGroup', 'TicketPrefix']

    for feature in categorical_features:
        if feature in combined_reset.columns:
            le = LabelEncoder()
            combined_reset[feature] = le.fit_transform(
                combined_reset[feature].astype(str)
            )

    # 2. –û–±—Ä–∞–±–æ—Ç–∫–∞ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    numeric_features = ['Age', 'Fare', 'SibSp', 'Parch', 'FamilySize']

    imputer = SimpleImputer(strategy='median')
    combined_reset[numeric_features] = imputer.fit_transform(
        combined_reset[numeric_features]
    )

    # 3. –°—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    scaler = StandardScaler()
    combined_reset[numeric_features] = scaler.fit_transform(
        combined_reset[numeric_features]
    )

    # 4. –í—ã–±–æ—Ä —Ñ–∏–Ω–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    selected_features = [
        'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked',
        'Title', 'FamilySize', 'IsAlone', 'Deck', 'AgeGroup', 'TicketLength'
    ]

    # –†–∞–∑–¥–µ–ª—è–µ–º –æ–±—Ä–∞—Ç–Ω–æ
    train_processed = combined_reset.iloc[:len(train_df)].copy()
    test_processed = combined_reset.iloc[len(train_df):].copy()

    train_processed['Survived'] = train_df['Survived'].values

    return train_processed, test_processed, selected_features
```

```python
def train_random_forest(X_train, y_train, X_val, y_val):
    """–û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ Random Forest —Å –æ—Ü–µ–Ω–∫–æ–π –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤."""
    print("\n" + "=" * 60)
    print("–û–ë–£–ß–ï–ù–ò–ï –ú–û–î–ï–õ–ò RANDOM FOREST")
    print("=" * 60)

    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–æ–¥–µ–ª–∏
    rf_model = RandomForestClassifier(
        n_estimators=N_ESTIMATORS,
        max_depth=10,
        min_samples_split=5,
        min_samples_leaf=2,
        max_features='sqrt',
        random_state=RANDOM_STATE,
        n_jobs=-1,
        verbose=0
    )

    # –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
    rf_model.fit(X_train, y_train)

    # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–∏
    y_pred = rf_model.predict(X_val)
    y_pred_proba = rf_model.predict_proba(X_val)[:, 1]

    # –û—Ü–µ–Ω–∫–∞ –º–æ–¥–µ–ª–∏
    accuracy = accuracy_score(y_val, y_pred)
    print(f"\n–¢–æ—á–Ω–æ—Å—Ç—å –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ: {accuracy:.4f}")

    print("\n–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
    print(classification_report(y_val, y_pred))

    # –ö—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è
    cv_scores = cross_val_score(
        rf_model, X_train, y_train, cv=5, scoring='accuracy'
    )
    print(f"\n–ö—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è (5 folds):")
    print(f"–°—Ä–µ–¥–Ω—è—è —Ç–æ—á–Ω–æ—Å—Ç—å: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")

    return rf_model, y_pred, y_pred_proba


def plot_feature_importance(model, feature_names):
    """–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤."""
    importance_df = pd.DataFrame({
        'feature': feature_names,
        'importance': model.feature_importances_
    }).sort_values('importance', ascending=False)

    print("\n" + "=" * 60)
    print("–í–ê–ñ–ù–û–°–¢–¨ –ü–†–ò–ó–ù–ê–ö–û–í:")
    print("=" * 60)
    print(importance_df.to_string(index=False))

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
    plt.figure(figsize=(12, 6))
    bars = plt.barh(
        importance_df['feature'][:15],
        importance_df['importance'][:15]
    )

    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –Ω–∞ –≥—Ä–∞—Ñ–∏–∫
    for bar in bars:
        width = bar.get_width()
        plt.text(
            width + 0.001,
            bar.get_y() + bar.get_height()/2,
            f'{width:.3f}',
            ha='left',
            va='center',
            fontsize=9
        )

    plt.xlabel('–í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∞')
    plt.ylabel('–ü—Ä–∏–∑–Ω–∞–∫')
    plt.title('–¢–æ–ø-15 –≤–∞–∂–Ω–µ–π—à–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (Random Forest)', fontsize=14)
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()

    return importance_df


def create_submission(model, test_data, features, test_df_original):
    """–°–æ–∑–¥–∞–Ω–∏–µ —Ñ–∞–π–ª–∞ –¥–ª—è –æ—Ç–ø—Ä–∞–≤–∫–∏ –Ω–∞ Kaggle."""
    # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    test_predictions = model.predict(test_data[features])

    # –°–æ–∑–¥–∞–Ω–∏–µ submission —Ñ–∞–π–ª–∞
    submission = pd.DataFrame({
        'PassengerId': test_df_original['PassengerId'],
        'Survived': test_predictions.astype(int)
    })

    # –ê–Ω–∞–ª–∏–∑ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print("\n" + "=" * 60)
    print("–ê–ù–ê–õ–ò–ó –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–ô –î–õ–Ø SUBMISSION:")
    print("=" * 60)
    survival_counts = submission['Survived'].value_counts()
    survival_percent = submission['Survived'].value_counts(normalize=True) * 100

    for i, (count, percent) in enumerate(zip(survival_counts, survival_percent)):
        status = "–í—ã–∂–∏–ª" if i == 1 else "–ü–æ–≥–∏–±"
        print(f"{status}: {count} ({percent:.1f}%)")

    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ñ–∞–π–ª–∞
    submission_filepath = 'submission.csv'
    submission.to_csv(submission_filepath, index=False)
    print(f"\n‚úÖ Submission —Ñ–∞–π–ª —Å–æ—Ö—Ä–∞–Ω–µ–Ω: {submission_filepath}")
    print(f"   –í—Å–µ–≥–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π: {len(submission)}")
    print(f"   –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ñ–æ—Ä–º–∞—Ç—É: {list(submission.columns)}")

    return submission
```
